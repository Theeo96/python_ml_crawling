```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

#### 데이터 준비

```python
df_iris = pd.read_csv('data/iris.csv').iloc[:,1:]
df_iris.head()
```


#### 데이터 탐색

```python
df_iris.info()
```

```python
# -----------------------------------
# 종속변수 빈도수 확인
#   불균형 여부 확인
# -----------------------------------

df_iris['Species'].value_counts()
```

```python
# -----------------------------------
# 독립변수
# -----------------------------------

features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

```python
# -----------------------------------
# 각 독립변수의 분포
# -----------------------------------
fig, ax = plt.subplots(2,2,figsize=(10,5), sharex=True, sharey=True)
axes = ax.flatten()
for i, feature in enumerate(features):
    sns.histplot(x=feature, data=df_iris, ax=axes[i])
    axes[i].set_title(feature)
plt.tight_layout()


```

```python
# -----------------------------------
# 독립변수와 종속변수간의 관계 시각화 -  boxplot
# -----------------------------------

fig, ax = plt.subplots(2,2,figsize=(10,5))
axes = ax.flatten()
for i, feature in enumerate(features):
    sns.boxplot(x='Species', y=feature, data=df_iris, ax=axes[i])
    axes[i].set_title(feature)
plt.tight_layout()
```

```python
sns.pairplot(df_iris)
```

```python
# -----------------------------------
# 독립변수간의 관계 시각화 - pariplot
# -----------------------------------

sns.pairplot(df_iris, hue='Species')
```
