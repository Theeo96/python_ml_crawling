#### 모든 페이지 데이터 수집

```python
key_decoding = 'hVVMVuTmCbA7r2MOOyDFItXgSn3eGaMpq9gp+a99j0J0f8Cr/zUi/1gGdnf44QBuGhMmcoFrmyluc1+QR7LhoQ=='
import pandas as pd
import requests
import json

dfs = []

# -----------------------------------
# 여러 페이지 데이터를 받기 위해 반복
# -----------------------------------
for i in range(1,14):
    
    # -----------------------------------
    # url 생성
    # -----------------------------------
    url = 'http://api.data.go.kr/openapi/tn_pubr_public_cltur_fstvl_api'
    params ={'serviceKey' : key_decoding, 
            'pageNo' : i, 
            'numOfRows' : '100', 
            'type' : 'json'}

    # -----------------------------------
    # url 생성
    # -----------------------------------            
    response = requests.get(url, params=params)

    # -----------------------------------
    # parsing
    #   json --> dict 
    # -----------------------------------
    data = response.json()['response']['body']['items']
    dfs.append(pd.DataFrame(data))
```

```python
df = pd.concat(dfs, ignore_index=True)
df.tail()
```
