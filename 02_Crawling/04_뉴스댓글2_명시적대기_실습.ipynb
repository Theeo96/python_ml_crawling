```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
import time
from datetime import datetime
import math
```

#### 웹드라이버 생성 및 페이지 이동

```python
url = "https://n.news.naver.com/article/005/0001707120"

driver = Chrome()
driver.get(url)

time.sleep(3)
```

#### 전체 댓글 수 추출

```python
# #comment_count
# review_total_cnt = driver.find_element(By.CSS_SELECTOR, "#comment_count").text
# review_total_cnt

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ------------------------------------------------------------------------------------
# 전체 댓글 수 요소가 DOM트리에 존재하고 보여지고 클릭 가능할 때까지 대기하고 요소를 리턴받는다.
# ------------------------------------------------------------------------------------

element = WebDriverWait(driver, 10)\
            .until(EC.element_to_be_clickable((By.CSS_SELECTOR, "#comment_count")))

# ------------------------------------------------------------------------------------
# 요소의 text가 '댓글'이 아닐 때까지 대기한다.
# ------------------------------------------------------------------------------------

WebDriverWait(driver, 10).\
        until(lambda d: d.find_element(By.CSS_SELECTOR, "#comment_count").text != "댓글")

# ------------------------------------------------------------------------------------
# 요소의 text 추출
# ------------------------------------------------------------------------------------

review_total_cnt = element.text
print(review_total_cnt)
```

#### 댓글 화면으로 이동

```python
driver.find_element(By.CSS_SELECTOR, "#comment_count").click()
time.sleep(3)
```

#### 더보기 클릭하여 모든 댓글 보이기

```python
# -----------------------------
# 더보기 클릭 횟수
#   더보기 클릭 할 때마다 20개의 댓글이 추가로 보여짐
#   ceil(전체 댓글 수/20)
# -----------------------------
pagecnt = math.ceil(int(review_total_cnt)/20)-1
# pagecnt = 3

# -----------------------------
# 더보기 클릭 
# -----------------------------

for i in range(pagecnt) :

    time.sleep(0.5)
    # driver.find_element(By.CSS_SELECTOR, ".u_cbox_btn_more").click()

    print(f'{i}번째 페이지 로드중')

    element = WebDriverWait(driver, 10) \
                .until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".u_cbox_btn_more")))
    element.click()

```

#### 데이터 추출

```python
# ------------------------------------------
# 댓글, 좋아요갯수, 싫어요갯수, 날짜
# ------------------------------------------

# 댓글 내용 : ".u_cbox_contents"
# 좋아요 갯수 : ".u_cbox_cnt_recomm"
# 싫어요 갯수 : ".u_cbox_cnt_unrecomm"
# 날짜 : ".u_cbox_date"
# 답글 수 : ".u_cbox_reply_cnt"
# 전체 블록 : ".u_cbox_area"

comment_list = []

comments = driver.find_elements(By.CSS_SELECTOR, ".u_cbox_area")

for comment in comments :
    try : review = comment.find_element(By.CSS_SELECTOR, ".u_cbox_contents").text
    except : review = np.nan
    
    try : recomm = comment.find_element(By.CSS_SELECTOR, ".u_cbox_cnt_recomm").text
    except : recomm = np.nan
    
    try : unrecomm = comment.find_element(By.CSS_SELECTOR, ".u_cbox_cnt_unrecomm").text
    except : unrecomm = np.nan

    try : date = comment.find_element(By.CSS_SELECTOR, ".u_cbox_date").text
    except : date = np.nan

    try : cnt = comment.find_element(By.CSS_SELECTOR, ".u_cbox_reply_cnt").text
    except : cnt = np.nan

    comment_list.append({"리뷰내용" : review,
                         "좋아요" : recomm,
                         "싫어요" : unrecomm,
                         "답글" : cnt,
                         "날짜" : date})

    
pd.DataFrame(comment_list)
```

#### 데이터 저장

```python
# -------------------------------
# 데이터프레임 생성
# -------------------------------

df = pd.DataFrame(comment_list)


# -------------------------------
# 파일로 저장
# -------------------------------

df.to_csv('data/나이키뉴스댓글.csv')
df.to_excel('data/나이키뉴스댓글.xlsx')
```
