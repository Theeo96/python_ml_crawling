#### 웹드라이버 객체 생성
- 웹드라이버 : 브라우저 조작을 위한 도구
- selenium의 webdriver 모듈에서 Chrome 브라우저를 제어하기 위한 클래스 제공

```python
# ------------------------------------------
# 크롬 웹드라이버 객체 생성
# ------------------------------------------

from selenium.webdriver import Chrome

driver = Chrome()
```

#### 브라우저 제어
    - `get(URL)` : 페이지로 이동하기기
    - `back()` : 뒤로가기
    - `forward()` : 앞으로가기
    - `refresh()` : 새로고침
    - `maximize_window()` : 브라우저크기 최대화
    - `minimize_window()` : 브라우저크기 최소화
    - `close()` : 웹드라이버로 생성한 현재 창 종료
    - `quit()` : 웹드라이버로 생성한 모든 창 종료

```python
# 브라우저 크기 최소화
driver.minimize_window()

# 브라우저 크기 최대화
driver.maximize_window()
```

```python
# 페이지 이동1
driver.get('http://www.naver.com')

# 페이지 이동2
driver.get('http://www.google.com')

# 페이지 이동3
driver.get('http://www.daum.net')
```

```python
# 이전 페이지로 이동
driver.back()
```

```python
# 다음 페이지로 이동
driver.forward()
```

```python
# 새로고침
driver.refresh()
```

```python
# 웹드라이버로 생성한 모든 창 종료
driver.quit()
```

#### 요소 가져오기
    - `find_element(요소선택방법, 선택자)` : 첫번째 요소 한개만만 가져오기
    - `find_elements(요소선택방법, 선택자)` : 요소 모두 가져오기 (리스트)

```python
# --------------------------------------
# 웹페이지 열고 이동하기 
# --------------------------------------

driver = Chrome()

url = 'http://www.naver.com'

driver.get(url)
```

```python
# --------------------------------------
# 검색창 선택하기
# --------------------------------------

from selenium.webdriver.common.by import By

search = driver.find_element(By.CSS_SELECTOR, '#query')
```

#### 요소 제어하기
    - `click()` : 요소 클릭
    - `send_keys()` : 키 입력

```python
# --------------------------------------
# 검색창 클릭
# --------------------------------------

search.click()
```

```python
# --------------------------------------
# 검색어 입력
# --------------------------------------

search.send_keys('인공지능')
```

```python
# --------------------------------------
# 엔터키 입력
# --------------------------------------

search.send_keys('\n')
```

```python
# --------------------------------------
# 브라우저 닫기
# --------------------------------------

driver.quit()
```
