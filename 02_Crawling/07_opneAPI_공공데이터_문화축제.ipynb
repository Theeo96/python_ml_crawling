#### 1페이지 데이터 수집

```python
key_decoding = 'hVVMVuTmCbA7r2MOOyDFItXgSn3eGaMpq9gp+a99j0J0f8Cr/zUi/1gGdnf44QBuGhMmcoFrmyluc1+QR7LhoQ=='
import pandas as pd
import requests
import json

# ------------------------------------------------
# url 생성 및 파라미터 정의
#   - opneAPI 활용 가이드에 따름
#   - 키 발급 
#   - 요청 url 생성
#   - 파라미터 정의
#   - json 타입으로 응답받도록 지정
# ------------------------------------------------
url = 'http://api.data.go.kr/openapi/tn_pubr_public_cltur_fstvl_api'

params ={'serviceKey' : key_decoding, 
         'pageNo' : '1', 
         'numOfRows' : '100', 
         'type' : 'json'}


# -----------------------------------
# 요청 및 응답
#   응답객체는 json 타입
# -----------------------------------    
response = requests.get(url, params=params)        
print(response.text)
print(type(response.text))

# -----------------------------------
# parsing
#   json --> dict 
# -----------------------------------
data = response.json()

# -----------------------------------
# 데이터 추출
# -----------------------------------
total_cnt = int(data['response']['body']['totalCount'])
data['response']['body']['items']

# -----------------------------------
# 페이지 계산
# -----------------------------------
page = int(total_cnt/100)+1
page
```

#### 모든 페이지 데이터 수집

```python
key_decoding = 'hVVMVuTmCbA7r2MOOyDFItXgSn3eGaMpq9gp+a99j0J0f8Cr/zUi/1gGdnf44QBuGhMmcoFrmyluc1+QR7LhoQ=='
import pandas as pd
import requests
import json

# -----------------------------------
# 여러 페이지 데이터를 받기 위해 반복
# -----------------------------------
dfs = []
for i in range(1, page+1):
    
    # -----------------------------------
    # url 생성
    # -----------------------------------
    url = 'http://api.data.go.kr/openapi/tn_pubr_public_cltur_fstvl_api'

    params ={'serviceKey' : key_decoding, 
            'pageNo' : i, 
            'numOfRows' : '100', 
            'type' : 'json'}

    # -----------------------------------
    # 요청 및 응답
    # -----------------------------------            
    response = requests.get(url, params=params)    

    # -----------------------------------
    # parsing
    #   json --> dict 
    # -----------------------------------
    data = response.json()
    data = data['response']['body']['items']
    dfs.append(pd.DataFrame(data))
```

```python
df = pd.concat(dfs, ignore_index=True)
df
```
