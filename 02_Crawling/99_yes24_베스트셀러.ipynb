```python
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
data = []
cnt = 100
i = 1

while True:
    
    time.sleep(np.random.uniform(1.0, 3.0))

    url = "https://www.yes24.com/product/category/bestseller"
    url += "?pageNumber="+str(i)
    url += "&pageSize=24"
    url += "&categoryNumber=001"

    response = requests.get(url)

    soup = BeautifulSoup(response.text, "html.parser")
    items = soup.select(".itemUnit .item_info")
    for item in items:
        title = item.select_one(".gd_name").text
        author = item.select_one(".authPub.info_auth").text.strip()
        pubdate = item.select_one(".authPub.info_date").text.strip()
        price = item.select_one("strong.txt_num").text.strip()
        try: grade = item.select_one(".rating_grade>.yes_b").text.strip()
        except: grade = None

        try: review_cnt = item.select_one(".rating_rvCount .txC_blue").text.strip()
        except: review_cnt = None
        data.append({"title":title, "author":author, "pubdate":pubdate, "price":price, "grade":grade, "review_cnt":review_cnt})
        if len(data)>=cnt:break     
    if len(data)>=cnt:break     
    i += 1   
        
print(review_cnt)
```

```python
df = pd.DataFrame(data)

df.info()
```
