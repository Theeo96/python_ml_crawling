```python
import requests
from bs4 import BeautifulSoup
import time
from datetime import datetime
import pandas as pd
import numpy as np
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
```

### url 지정 (멜론)

```python
url = 'https://www.melon.com/chart/'

driver = Chrome()
driver.get(url)

wait = WebDriverWait(driver, 10)

time.sleep(2)
```

### 크롤링 수행

```python
# 곡 순위 : "td:nth-child(2) .rank" (맨 위에 '순위' 문자열 포함해서 2번부터 지정)
# 곡 제목 : ".ellipsis.rank01 a"
# 아티스트 이름 : ".ellipsis.rank02 a"
# 앨범 제목 : ".ellipsis.rank03 a"
# 좋아요 수 : ".button_etc.like .cnt"

# 곡 블록 : "tbody tr.lst50, tbody tr.lst100" (100은 51위부터 시작)

song_list = []

songs = driver.find_elements(By.CSS_SELECTOR, "tbody tr.lst50, tbody tr.lst100")

for song in songs : 

    rank = song.find_element(By.CSS_SELECTOR, "td:nth-child(2) .rank").text

    title = song.find_element(By.CSS_SELECTOR, ".ellipsis.rank01 a").text
    
    artist = song.find_element(By.CSS_SELECTOR, ".ellipsis.rank02 a").text.strip()

    album = song.find_element(By.CSS_SELECTOR, ".ellipsis.rank03 a").text.strip()

    like_count = song.find_element(By.CSS_SELECTOR, ".button_etc.like .cnt").text

    song_list.append({"순위" : rank,
                      "곡 제목" : title,
                      "아티스트 명" : artist,
                      "앨범 명" : album,
                      "좋아요" : like_count})

```

### 데이터프레임으로 만들기

```python
df = pd.DataFrame(song_list).reset_index(drop=True)
df
```

### 데이터 저장

```python
df.to_csv('data/멜론차트크롤링.csv', index=False)
df.to_excel('data/멜론차트크롤링.xlsx')
```
